```python
# write a test to check that all elements along the diagonal of a matrix are 1
import numpy as np
def test_diagonal_one(m):
    return np.allclose(np.diag(m), np.ones(m.shape[0]))


# test that the matrix piecewise multiplied by its transpose gives a matrix of ones
# write documentation for the function below
def test_matrix_mult(m):
    
    return np.allclose(m * m.T, np.ones(m.shape))


# test the above two functions on the matrices in the matrices list
for m in matrices:
    print(m*m.T)
    print(test_diagonal_one(m))
    print(test_matrix_mult(m))
```

```python
# write a generic class name currency_holding that has as attributes: month, currency, amount, transaction history; and @property decorators for each attribute. Also include documentation.
class CurrencyHolding:
    """
    The currency holding class represents a currency holding for a given month.
    We only keep the path leading to the largest currency holding.
    We keep the transaction history in a list of tuples (from_currency, to_currency, rate, amount).
    """
    def __init__(self, month, currency, amount, transaction_history):
        """
        The constructor for the currency holding class.
        """
        self.month = month
        self.currency = currency
        self.amount = amount
        self.transaction_history = transaction_history

    
    def __str__(self):
        """
        The string representation of the currency holding.
        """
        return f'Month {self.month} Currency {self.currency} Amount {self.amount} \n Transaction History {self.transaction_history}'

    def __repr__(self):
        """
        The string representation of the currency holding.
        """
        return f'CurrencyHolding(month={self.month}, currency={self.currency}, amount={self.amount}, transaction_history={self.transaction_history})'
    
    @property
    def month(self):
        """
        Get the month.
        """
        return self._month

    @month.setter
    def month(self, month):
        """
        Set the month.
        """
        self._month = month

    @property
    def currency(self):
        """
        Get the currency.
        """
        return self._currency

    @currency.setter
    def currency(self, currency):
        """
        Set the currency.
        """
        self._currency = currency

    @property
    def amount(self):
        """
        Get the amount.
        """
        return self._amount

    @amount.setter
    def amount(self, amount):
        """
        Set the amount, making sure amount is convertible to float and positive.
        """
        try:
            amount = float(amount)
        except ValueError:
            raise ValueError('Amount must be convertible to float.')
        if not amount>=0:
            raise ValueError('Amount must be positive.')
        self._amount = amount

    @property
    def transaction_history(self):
        """
        Get the transaction history.
        """
        return self._transaction_history

    @transaction_history.setter
    def transaction_history(self, transaction_history):
        """
        Set the transaction history.
        """
        self._transaction_history = transaction_history
```

```python
# define a function that multiplies element-wise a vector by a matrix such that the result is a matrix, make sure the vector is reshaped to a column vector and write a test for the function using the example in the cell above, and write documentation for the function. test the function
def multiply_vector_matrix(vector, matrix):
    """
    Multiply a vector by a matrix element-wise.
    And performs a test to make sure the input is a vector and the output is a matrix.
    """
    if not len(vector.shape) == 1:
        raise TypeError('Vector must be a 1D array.')
    if not len(matrix.shape) == 2:
        raise TypeError('Matrix must be a 2D array.')
    if not vector.shape[0] == matrix.shape[0]:
        raise ValueError('The number of rows in the vector must be equal to the number of rows in the matrix.')
    return np.einsum('i,ij->ij', vector, matrix)


def test_multiply_vector_matrix():
    """
    Test the multiply_vector_matrix function.
    """
    a = np.array([1.1,0.9])
    b = np.array([[1, 1.1],[1.5,1]])
    assert np.all(multiply_vector_matrix(a, b) == a.reshape((2,1))*b)

    

```

```python
# re-write the code in the cell above in a object oriented way, suppose we have the class already defined
import numpy as np
class CurrencyHolding:
    """
    The currency holding class represents a currency holding for a given month.
    We only keep the path leading to the largest currency holding.
    We keep the transaction history in a list of tuples (from_currency, to_currency, rate, amount).
    """
    def __init__(self, currency, order, amount, transaction_history):
        """
        The constructor for the currency holding class.
        """
        self._order = order
        self.currency = currency
        self.amount = amount
        self.transaction_history = transaction_history

    
    def __str__(self):
        """
        The string representation of the currency holding.
        """
        return f'Currency {self.currency} Amount {self.amount} \n Transaction History {self.transaction_history}'

    def __repr__(self):
        """
        The string representation of the currency holding.
        """
        return f'CurrencyHolding(currency={self.currency}, amount={self.amount}, transaction_history={self.transaction_history})'
    
    # @property
    # def month(self):
    #     """
    #     Get the month.
    #     """
    #     return self._month

    # @month.setter
    # def month(self, month):
    #     """
    #     Set the month.
    #     """
    #     self._month = month

    @property
    def currency(self):
        """
        Get the currency.
        """
        return self._currency

    @currency.setter
    def currency(self, currency):
        """
        Set the currency.
        """
        self._currency = currency


    @property
    def order(self):
        """
        Get the order.
        """
        return self._order
    
    
    @order.setter
    def order(self, order):
        """
        Set the order.
        """
        raise ValueError(f'Order is {order} and cannot be changed.')


    @property
    def amount(self):
        """
        Get the amount.
        """
        return self._amount

    @amount.setter
    def amount(self, amount):
        """
        Set the amount, making sure amount is convertible to float and positive.
        """
        try:
            amount = float(amount)
        except ValueError:
            raise ValueError('Amount must be convertible to float.')
        if not amount>=0:
            raise ValueError('Amount must be positive.')
        self._amount = amount

    @property
    def transaction_history(self):
        """
        Get the transaction history.
        """
        return self._transaction_history

    @transaction_history.setter
    def transaction_history(self, transaction_history):
        """
        Set the transaction history.
        """
        self._transaction_history = transaction_history

    # def add_transaction(self, transaction):
    #     """
    #     Add a transaction to the transaction history.
    #     """
    #     self._transaction_history.append(transaction)



# Define a class for the simulation
class Simulation:
    """
    The simulation class represents a simulation of a system.
    """
    def __init__(self, exchange_rates_list, currency_order):

        print('Initialising simulation')
        # We have as many currency holdings as there are currencies
        self.currency_holdings = [CurrencyHolding(
            currency, order, 1 if currency == 'GBP' else 0, []) for order, currency in enumerate(currency_order)]
        self.number_months = len(exchange_rates_list)
        self.exchange_rates_list = exchange_rates_list
        self.month = 0
        self.current_exchange_rate = exchange_rates_list[0]
        # Compile a numpy list of current holding amounts in order of currency_order
        self.current_holdings = np.array(
            [currency_holding.amount for currency_holding in self.currency_holdings])

        print(self.current_holdings)

    # Define a method to perform one iteration of the simulation
    def step(self):

        # Print the current month, current holdings, and exchange rates
        print('Month', self.month)
        print('Current holdings', self.current_holdings)
        print('Exchange rates', self.exchange_rates_list[self.month])

        # Get the transaction results
        transaction_results = multiply_vector_matrix(self.current_holdings, self.exchange_rates_list[self.month])
        print('Transaction results', transaction_results)

        # The crucial step is that after the 16 transactions take place, we are only interested in the ones that have the maximum possible holding for each currency
        # Get the maximum possible holding for each currency, i.e. the maximum along each column
        max_holdings = np.max(transaction_results, axis=0)
        # Get the indices of the maximum possible holding for each currency, i.e. for each column, get the index of the maximum - that gives you the currency holding the maximum came from
        max_indices = np.argmax(transaction_results, axis=0)
        # Now that we have the indices in the currency_order list, we can get the holding account the maximum came from. The holding account is defined by the currency. 
        source_currency_holdings = [self.currency_holdings[max_indices[i]]
                        for i in range(len(max_indices))]
        print('Max holdings', max_holdings)
        print('Max indices', max_indices)
        print('Max currencies', [currency_holding.currency for currency_holding in source_currency_holdings])

        # We save a copy of the current state of the transaction histories
        previous_transation_histories = [list(currency_holding.transaction_history) for currency_holding in self.currency_holdings]
        print('Previous transaction histories', previous_transation_histories)

        # Update the currency holdings.
        """
        For each currency holding, after each transaction round there are 2 posibilities:
            1. The most currency was attained by not doing an exchange, i.e. keeping the same currency.
                In this case, we just append the new transaction to the transaction history of the current holding.

            2. The most currency was attained by doing an exchange from another, source, currency.
                In this case, we delete the transaction history of the current holding and instead branch off from the transaction history of the source currency holding.

        This is because, for a given currency holding, we are only interested in the transaction history that resulted in the maximum amount of currency.
        """
        for current_currency_holding, new_amount, source_currency_holding in zip(self.currency_holdings, max_holdings, source_currency_holdings):
            print('Currency', current_currency_holding.currency)
            print('Amount', new_amount)
            print('Max currency', source_currency_holding.currency)

            # Update the amount of currency held
            current_currency_holding.amount = new_amount

            # Create the new transaction entry for the transaction history
            new_transaction = (self.month, source_currency_holding.currency, current_currency_holding.currency,
                               self.exchange_rates_list[self.month][source_currency_holding.order, current_currency_holding.order], new_amount)

            # If the transaction was an exchange between currencies, then we branch off the transaction history of the source currency holding
            if current_currency_holding.currency != source_currency_holding.currency:
                current_currency_holding.transaction_history = list(previous_transation_histories[
                    source_currency_holding.order]) 

            # Update the transaction history with the new transaction
            current_currency_holding.transaction_history.append(
                new_transaction)
            print('Transaction history', current_currency_holding.transaction_history)
       
        # Update the month
        self.month += 1
       
        # Update the current holdings
        self.current_holdings = np.array([currency_holding.amount for currency_holding in self.currency_holdings])


    
    # Define a method to perform the last iteration of the simulation, 
    # where we imagine all currency holdings are converted to GBP at the end of the simulation, and we pick the best performing one.
    def last_step(self):

        # Print the current month, current holdings, and exchange rates
        print('Month', self.month)
        print('Current holdings', self.current_holdings)
        print('Exchange rates', self.exchange_rates_list[self.month])

        # get the output matrix
        max_holdings_in_gbp = self.current_holdings * \
            self.exchange_rates_list[self.month][:,
                                                 currency_order.index('GBP')]
        
        for source_currency_holding, max_GBP_amount in zip(self.currency_holdings, max_holdings_in_gbp):
            print('Currency', source_currency_holding.currency)
            print('Amount ', max_GBP_amount)
            print('Final currency GBP')

            # Update the transaction histories
            source_currency_holding.amount = max_GBP_amount
            new_transaction = (self.month, source_currency_holding.currency, 'GBP', self.exchange_rates_list[self.month][source_currency_holding.order, currency_order.index('GBP')], max_GBP_amount)
            source_currency_holding.transaction_history.append(new_transaction)

            print('Final transaction history', source_currency_holding.transaction_history)


        # Update the current holdings
        self.current_holdings = np.array([currency_holding.amount for currency_holding in self.currency_holdings])

        # Update the month
        self.month += 1



        holding_with_best_transaction_path = max(self.currency_holdings, key=lambda x: x.amount)
        print('')
        print('The best transaction path we could taken given the matrices of exchange rates is: ', holding_with_best_transaction_path.transaction_history)
        print('The final amount of GBP we would have is: ', holding_with_best_transaction_path.amount)

    # Define a method to run the simulation
    def run(self):
        # Perform multiple iterations of the simulation
        for month in range(self.number_months-1):
            # Call step method
            self.step()
            # Print current state
            print(f"-----------------------------Month {month} complete.-----------------------------")

        # Run the last step
        self.last_step()

        # except Exception as e:
        #     # Handle any errors
        #     print(f"Simulation failed: {e}")

    # Define a static method to read in the data
    @staticmethod
    def read_data(file_name,n_months=12):
        # Read in the data

        # Open the file and delete any new line characters
        with open(file_name, 'r') as f:
            data = f.read().replace('\n', '').replace('[', '').replace(']', '').replace(' ','').split(',')
            # Convert the data to a numpy array of floats
            data = np.array(data,dtype=float)
            # Reshape the data to the correct shape
            data = data.reshape((n_months, 4, 4))
            # Return the data
            return data

def multiply_vector_matrix(vector, matrix):
    """
    Multiply a vector by a matrix element-wise.
    And performs a test to make sure the input is a vector and the output is a matrix.
    """
    if not len(vector.shape) == 1:
        raise TypeError('Vector must be a 1D array.')
    if not len(matrix.shape) == 2:
        raise TypeError('Matrix must be a 2D array.')
    if not vector.shape[0] == matrix.shape[0]:
        raise ValueError(
            'The number of rows in the vector must be equal to the number of rows in the matrix.')
    return np.einsum('i,ij->ij', vector, matrix)


def test_multiply_vector_matrix():
    """
    Test the multiply_vector_matrix function.
    """
    a = np.array([1.1, 0.9])
    b = np.array([[1, 1.1], [1.5, 1]])
    result = np.array([[1.1, 1.21], [1.35, 0.9]])
    # check 2 2D arrays are equal element-wise within machine precision
    assert np.allclose(multiply_vector_matrix(a, b), result)

if __name__ == '__main__':

    # Read in the data
    data_file = 'currency_data.txt'
    n_months = 12 # Make sure this is consistent with the data file
    currency_order = ['GBP', 'USD', 'EUR', 'JPY']# This should also be consistent with the data file

    # Read in exchange rates matries from the data file
    exchange_rates_list = Simulation.read_data(data_file,n_months=n_months)


    assert exchange_rates_list.shape[1:][0] == len(currency_order), 'The shape of the exchange rates matrix is not consistent with the number of currencies.'
    assert exchange_rates_list.shape[1:][1] == len(currency_order), 'The shape of the exchange rates matrix is not consistent with the number of currencies.'

    # Create a simulation instance
    sim = Simulation(exchange_rates_list, currency_order)
    # Run the simulation
    sim.run()
```

```python
def multiply_vector_matrix(vector, matrix):
    """
    Multiply a vector by a matrix element-wise.
    And performs a test to make sure the input is a vector and the output is a matrix.
    """
    if not len(vector.shape) == 1:
        raise TypeError('Vector must be a 1D array.')
    if not len(matrix.shape) == 2:
        raise TypeError('Matrix must be a 2D array.')
    if not vector.shape[0] == matrix.shape[0]:
        raise ValueError(
            'The number of rows in the vector must be equal to the number of rows in the matrix.')
    return np.einsum('i,ij->ij', vector, matrix)


def test_multiply_vector_matrix():
    """
    Test the multiply_vector_matrix function.
    """
    a = np.array([1.1, 0.9])
    b = np.array([[1, 1.1], [1.5, 1]])
    result = np.array([[1.1, 1.21], [1.35, 0.9]])
    # check 2 2D arrays are equal element-wise within machine precision
    assert np.allclose(multiply_vector_matrix(a, b), result)

test_multiply_vector_matrix()
```
